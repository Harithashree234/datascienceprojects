# DATA VISUALIZATION IN PYTHON

In Python, popular libraries like Matplotlib, Seaborn, and Plotly are used for creating various visualizations. These visualizations are often classified into two categories: 

 * Univariate Analysis and
 * Bivariate Analysis.

#### a) Univariate Analysis:
   Univariate analysis refers to the analysis of a single variable or feature in a dataset. The main goal of univariate analysis is to understand the distribution, central tendency, spread, and shape of the data for one feature at a time.

#### b) Bivariate Analysis:
   Bivariate analysis, on the other hand, focuses on the relationship between two variables. The aim is to explore how two variables interact with each other, whether they are correlated, and if there is any pattern or trend that links them.

# 5. UNIVARIATE ANALYSIS:

### i) HISTOGRAM:

~ Histogram shows the frequency distribution of a single numerical variable by grouping data into bins. 

~ It's used to see the overall shape (e.g., normal, skewed) of the data.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['beds'],  bins=30, color="blue")
plt.title("Distribution of beds", fontsize=14)
plt.xlabel("Beds", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

##### Chart Inferences:

1. The majority of listings have between 0 to 2 beds, with a significant peak at 1 bed.
2. Listings with more than 3 beds are much less common, and their frequencies drop sharply as the number of beds increases.
3. This suggests that most accommodations cater to smaller groups or individuals, which is typical for urban rental markets or listings focused on solo travelers or couples.

#

### ii) KDE PLOT (Kernel Density Estimate):

~ A KDE plot (Kernel Density Estimate plot) is a way to visualize the probability density function (PDF) of a continuous random variable. 

~ It is a smooth, continuous estimate of the underlying distribution of the data, providing a more refined and smoother version of the histogram.

```python
# To fix the plot size
plt.figure(figsize=(10,6))

# Create KDE plot for review_scores_rating
sns.kdeplot(data=df, x='review_scores_rating', hue='room_type', palette=['red','blue','green'], shade=True, legend=True)

# Title and labels
plt.title('KDE Plot of review for different types of room')
plt.xlabel('Values')
plt.ylabel('Density')

# Show plot
plt.show()
```

##### Chart Inferences:

1. Entire home/apt (Red): This category shows a high concentration of review scores near 100, indicating that these listings are frequently rated highly.
2. Private room (Blue): These listings also tend to have high review scores, although the density is slightly lower than "Entire home/apt" in the highest score range.
3. Shared room (Green): Shared rooms have a lower density overall, but the review scores are distributed similarly to the other categories, with a peak near 100.

#

### iii) BOX PLOT:

A box plot (also known as a box-and-whisker plot) is a great way to visualize the distribution of a dataset and to highlight key statistics such as the median, quartiles, and outliers. 

~ The lower edge of the box represents the first quartile (Q1) (25th percentile).

~ The upper edge represents the third quartile (Q3) (75th percentile).

~ The line inside the box represents the median (50th percentile).

```python
plt.figure(figsize=(8, 5))

# Create the univariate box plot for log_price
sns.boxplot(data=df, y='log_price', color='cyan')

# Adjusted positions for the five-number summary for 'log_price'
plt.text(x=-0.1, y=df['log_price'].min(), s='min', color="black", fontsize=10)
plt.text(x=-0.1, y=df['log_price'].quantile(0.25), s='Q1', color="black", fontsize=10)
plt.text(x=-0.1, y=df['log_price'].median(), s='median(Q2)', color="black", fontsize=10)
plt.text(x=-0.1, y=df['log_price'].quantile(0.75), s='Q3', color="black", fontsize=10)
plt.text(x=-0.1, y=df['log_price'].max(), s='max', color="black", fontsize=10)

# Annotate the outliers (adjust xy values accordingly)
plt.annotate('Outliers', xy=(0, df['log_price'].max() - 1), xytext=(-0.3, df['log_price'].max() - 2),
             arrowprops=dict(facecolor='black', arrowstyle='simple'))

# Title and labels
plt.title('Box Plot of Log Price', fontsize=14)
plt.ylabel('Log Price', fontsize=12)

# Show plot
plt.show()

```

##### Chart Inferences:

1. The box plot provides insight into the distribution of log prices, showing a right-skewed distribution with a broad range of values and a few outliers in the high-price range. 
2. It highlights that most listings are priced within a moderate range, while the few outliers suggest the presence of premium listings or extreme values.

```python
plt.figure(figsize=(8, 5))

# Create the univariate box plot for log_price
sns.boxplot(data=df, y='accommodates', color='cyan')

# Adjusted positions for the five-number summary for 'log_price'
plt.text(x=-0.1, y=df['accommodates'].min(), s='min', color="black", fontsize=10)
plt.text(x=-0.1, y=df['accommodates'].quantile(0.25), s='Q1', color="black", fontsize=10)
plt.text(x=-0.1, y=df['accommodates'].median(), s='median(Q2)', color="black", fontsize=10)
plt.text(x=-0.1, y=df['accommodates'].quantile(0.75), s='Q3', color="black", fontsize=10)
plt.text(x=-0.1, y=df['accommodates'].max(), s='max', color="black", fontsize=10)

# Annotate the outliers (adjust xy values accordingly)
plt.annotate('Outliers', xy=(0, df['accommodates'].max() - 1), xytext=(-0.3, df['accommodates'].max() - 2),
             arrowprops=dict(facecolor='black', arrowstyle='simple'))

# Title and labels
plt.title('Box Plot of accommodates', fontsize=14)
plt.ylabel('accommodates', fontsize=12)

# Show plot
plt.show()
```

##### Chart Inferences:

1. The box plot provides insight into the distribution of accommodates, showing a right-skewed distribution with a broad range of values and a few outliers in the high-accommodation range.
2. Most listings accommodate between 24 people, with outliers going up to 16.

### iv) PIE CHART:

~ A pie chart is used to display the relative frequencies of different categories in a single variable. 

~ The size of each slice in the pie chart is proportional to the percentage of observations in each category. 

~ This means a larger slice represents a category that occurs more frequently in the dataset, while a smaller slice represents a less frequent category.

```python
# summarize the frequency of each dismissal mode
dismissal_counts=df['cancellation_policy'].value_counts()

# create a pie chart
plt.figure (figsize=(6,6))
plt.pie(dismissal_counts, 
        labels=dismissal_counts.index, # label name 
        autopct='%1.1f%%', # decimal points
        startangle= 80, # determines the pie for first value
        colors=sns.color_palette("husl"))
        
# create title for the chart
plt.title("Distribution of Cancellation policy", fontsize=12)

# show the chart
plt.show()
```

##### Chart Inferences:

 Strict policies dominate the distribution, while flexible and moderate policies are less common but still significant. This suggests that property managers or hosts may prioritize stricter rules to reduce last-minute cancellations.

### v) COUNT PLOT:

~ A count plot is a bar plot that shows the frequency (count) of categories in a categorical variable. 

~It is commonly used to visualize how many observations belong to each category in a dataset.

```python
# Plot the count plot for 'room_type'
sns.countplot(data=df, x='instant_bookable', palette='Set1')

# Get the current axes - gca()
ax = plt.gca()

# Add labels on top of the bars
for bar in ax.patches:  # Iterates through each bar
    # Get the bar's x-coordinate, width, and height
    x = bar.get_x() + bar.get_width() / 2  # center of the bar
    y = bar.get_height()  # height of the bar
    ax.text(x, y, 
            f'{y:.0f}',  # Display the count as an integer
            ha='center', va='bottom', fontsize=10)

# Customize and print the chart
plt.title("Frequency of Instant Booking", fontsize=14)
plt.xlabel("Instant bookable", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

##### Chart Inferences:

The count plot reveals the frequency of each instant bookings in the dataset, showing which type is most common (e.g., "Entire home/apt" may dominate) and which is least common (e.g., "Shared room" might be rare).

### i) SCATTER PLOT:

~ A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph, with one variable plotted on the x-axis and another on the y-axis. 

~ It is used to identify relationships, trends, and correlations between two continuous variables.

```python
plt.figure(figsize=(8, 5))

# Scatter plot with customized color and style
sns.scatterplot(x='accommodates', y='log_price', data=df, 
                hue='room_type', palette=['salmon','crimson','orange'], marker='s', s=50, edgecolor='black')

# Title and labels
plt.title('Scatter Plot of Log Price vs. Accommodates', fontsize=14)
plt.xlabel('Number of Guests Accommodated', fontsize=12)
plt.ylabel('Log Price', fontsize=12)

# Show plot
plt.show()
```

##### Chart Inferences:

1. There appears to be a positive relationship between the number of guests a listing can accommodate and its log price. 
2. As the number of guests increases, the price of the listing tends to be higher, though there is significant variability.
3. The price distribution varies across room types, with Entire home/apt having more spread and higher prices.

### ii) BAR CHART:

~ A bar plot (or bar chart) is a data visualization that represents categorical data with rectangular bars, where the length of each bar is proportional to the value of the category it represents. 

~It is used to compare quantities across different categories or groups.

```python
# Average log_price by room_type and instant_bookable
mixed_data = df.groupby(['room_type', 'instant_bookable'])['log_price'].mean().unstack()

# Plot the mixed analysis
plt.figure(figsize=(10, 6))
mixed_data.plot(kind='bar', color=['cyan', 'lime'], edgecolor='black', width=0.8)
plt.xlabel('Room Type')
plt.ylabel('Average Log Price')
plt.title('Average Log Price by Room Type and Instant Bookable')
plt.xticks(rotation=45)
plt.legend(title='Instant Bookable', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

#####  Chart Inferences:

1. Entire home/apt has the highest average log price, with instant bookable listings slightly more expensive than non-instant ones.  
2. Private room shows a noticeable price gap, where instant bookable listings are priced higher than non-instant ones.  
3. Shared room has the lowest average log price, with little difference between instant and non-instant bookable options.  
4. Instant bookable listings generally have higher prices across all room types, indicating a premium for convenience.

### iii) LINE PLOT:

~ A line plot is a graph that displays data points connected by straight lines, commonly used to show trends or changes over a continuous variable. 

~ It is particularly useful for visualizing the relationship between two variables, especially when one is time-dependent or ordered.


```python
# Plot the line chart
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x='bedrooms', y='review_scores_rating', marker='o', color='magenta', errorbar=None)

# customise the chart
plt.title('Review Scores Rating vs. Number of Bedrooms', fontsize=12)
plt.xlabel('Number of Bedrooms', fontsize=12)
plt.ylabel('Review Scores Rating', fontsize=12)

# print the chart
plt.show()
```

### iv) Violin plot:

~ A violin plot combines aspects of both box plots and kernel density plots, providing a visualization of the distribution and probability density of a variable. 

~ It shows the datas distribution across different categories, highlighting the spread, median, and quartiles of the data.

```python
# Create the violin plot for log price by room type
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='room_type', y='log_price', palette=['cyan','magenta','green'], inner='box')

# Customize the chart
plt.title('Log Price Distribution by Room Type', fontsize=14)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Log Price', fontsize=12)

# Display the chart
plt.show()
```

##### Chart Inferences:

1. Entire homes/apts have the highest median log price with less variability, indicating they are the costliest and most consistent in pricing.
2. Private rooms show moderate prices with slightly more variability, acting as a middle ground between entire homes and shared rooms.
3. Shared rooms have the lowest median log price but the widest spread, reflecting affordability with some high-priced outliers.
4. Price distributions show entire homes are concentrated, while shared rooms have more extreme variations.

# MULTI VARIATE ANALYSIS:

### i) Heatmap:

~ A heatmap is a graphical representation of data where individual values contained in a matrix are represented with varying colors. 

~ It's often used to visualize the correlation between numerical variables or compare values across different categories.

```python
# Heatmap for correlations between numerical variables
plt.figure(figsize=(8, 6))
correlation_matrix = df[['log_price', 'accommodates', 'bedrooms', 'beds', 'review_scores_rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

##### Chart Inferences:

1. Log Price is positively correlated with accommodates, bedrooms, and beds, indicating larger listings are priced higher.  
2. Accommodates has a strong correlation with bedrooms and beds, as bigger spaces naturally house more people.  
3. Review Scores Rating shows little to no correlation with other variables, suggesting guest satisfaction is independent of listing size or price.  
4. Bedrooms and Beds are highly correlated, reflecting that listings with more bedrooms usually provide more sleeping arrangements.
